9. Розбити масив по вертикалі на 2 рівні частини (не використовувати абсолютні числа, мають бути два массиви по 4 колонки)

```python
top, bottom  = np.split(d2d, 2, axis = 0)

print("Shape top part = ", top.shape)
print("Shape bottom part = ", bottom.shape)
print("Top part = ", top)
print("Bottom part =", bottom)
```

10. Відсортувати обидва массиви по 1-й колонці: 1-й за збільшенням, 2-й за зменшенням

```python
top = top[np.argsort(top[:, 0])]
bottom = bottom[np.argsort(bottom[:, 0])[::-1]]
print("Sorted top: ", top)
print("Sorted bottom", bottom)
```

11. Зібрати обидва массиви в одне ціле

```python
combine = np.vstack((top, bottom))
print(combine)
```

12. Знайти найбільш часто повторюване значення в массиві

```python
flatten_combine = combine.flatten()
unique_values, counts = np.unique(flatten_combine, return_counts=True)
most_frequent_index = np.argmax(counts)
print("Most freq value = ", unique_values[most_frequent_index])
```

13. Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.

```python
def fn(data, column_index):
    mean_value = np.mean(data[:, column_index])
    data[:, column_index] = np.where(
        data[:, column_index] < mean_value,
        data[:, column_index] * 2,
        data[:, column_index] / 4
    )
    
    return data
```

14. Застосувати отриману функцію до 3-ї колонки

```python
fn(combine, 2)
```
